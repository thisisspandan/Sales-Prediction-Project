## 3. Data Processing and Handling

Before we begin our actual coding, we would like to explain that we are going to use a Top-Down approach to our data model unlike a conventional Bottom-Up approach.

In conventional Bottom-Up approach:

* We start with data cleaning
* Perform Exploratory Data Analysis
* Feature Engineering
* Data Augmentation
* Train Our Model
* Make Predictions

What we are trying to do rather involves a different order of things:

* We start off with some trivial data cleaning
* Train Our Model
* Understand Our Model
* Making the Model Better by methods like Hyper-Paramter Tuning
* Extracting Features
* Removing Rendundant Features
* Establishing Relationships between Features
* Re-Training Our Model 
* Make Predictions



For our Top - Down Approach of dealing with our dataset, we chose Random Forest ensemble learning method for:

Most data science approaches throw at us a large number of dozen algorithms to learn the data, with a brief technical descriprtion with a toy example. 

The good news is that the modern day data science learning methods can be distilled or reduced to a couple of key techniques that have a very high applicability. Recent studies showed us that majority of datasets can be best modelled with just two methods:

* Ensembles of Decision Trees suited for structured data (i.e. Random Forests and Gradient Boosting Machines)
* Multi Layered Neural Networks learnt with Stochastic Gradient Descent(SGD) suited for unstructured data (i.e for Shallow/Deep Learning)

We chose Random Forest Ensemble in particular because:

* It can predict anything of anykind i.e continous ( Regression ) or category ( Classification )
* It can predict anything of anytype i.e. structured or unstructured
* It doesn\`t overfit too badly, and it is very easy to stop it from overfitting
* It doesn\`t require a separate validation set, it can genralize well on a single dataset.
* It doen\`t require too many statistical assumptions, if not very less i.e it doesn\`t assume the data to be normalized.
* It requires almost a very few pieces of feature engineering.

## 4. Building our Model

### Building our Random Forest

Lets try building our model with separate training and validation set

```
def rmse(x,y): return math.sqrt(((x-y)**2).mean())

def print_score(m):
    res = [rmse(m.predict(X_train), y_train), rmse(m.predict(X_valid), y_valid),
                m.score(X_train, y_train), m.score(X_valid, y_valid)]
    if hasattr(m, 'oob_score_'): res.append(m.oob_score_)
    print(res)
```

```
m = RandomForestRegressor(n_jobs=-1)
%time m.fit(X_train, y_train)
print_score(m)
```

We can see that our model has a high r2 value, but we can notice from our validation set score that we are overfitting quite badly. To understand what is happening, let us simplify a couple of things.

### Introducing Bootsraps & Bagging

We mentioned earlier that Random Forest is a collection of Decision Trees. What it means is that Random Forest is a bagging of several different models, where we average insights from each of these models. This way of averaging insights is called **Ensembling**.

What we do is pass a random subset (bootstrap) of our dataset to each of the different models and train our models on it. Since the datasets are random, each model provides different insights as possible and we make it such a way the errors are not correlated as much as possible.

We call this individual model or Decision tree by the name Estimator.


```
# Lets see what our base model actually does
# By default the number of estimators for a random forest is 10

m = RandomForestRegressor(n_jobs=-1)
%time m.fit(X_train, y_train)
print_score(m)
```

```
# let us now look at the predictions of each of the trees on a particular point 
preds = np.stack([t.predict(X_valid) for t in m.estimators_])
preds[:,0] , np.mean(preds[:,0]) ,y_valid[0]
```

Let us now see how our model performs upon increasing the number of trees or estimators

```
plt.plot([metrics.r2_score(y_valid,np.mean(preds[:i+1], axis=0)) for i in range(10)])
```

```
for i in [20,40,80]:
  m = RandomForestRegressor(n_estimators=i,n_jobs=-1)
  m.fit(X_train,y_train)
  i, print_score(m)
```

#### Out of Bag (OOB) Score

In order to see how well our model generalizes on the training set instead of the validation set. OOB score is measured on the rows of data which were not included random sample of dataset (bootstrap) passed onto the estimators. OOB score is the mean of the scores of all estimators.



```
m = RandomForestRegressor(n_estimators=40, n_jobs=-1, oob_score=True)
%time m.fit(X_train,y_train)
print_score(m)
```

### Reducing Overfitting

#### Sub-sampling

Here instead of choosing a random 30000 subset and choosing 20000 data points from the subset, we now choose random 20000 data points from the dataset and pass it to every estimator. i.e each estimator has a different subset of training datapoints. Thereby making every estimator much less correlated to another estimator and speed things up.



```
set_rf_samples(20000)
```

```
# We need to set oob_score=False since it is an out of box method
m = RandomForestRegressor(n_estimators=40,n_jobs=-1,oob_score=False)
%time m.fit(X_train,y_train)
print_score(m)
```

#### Tree Building Parameters

In our decision tree based model, we can limit our model capability to learn by making few modifications to how the tree grows in terms of learning.

Some of the parameters that help us in modifying the tree are as follows:

**min_sample**

min_sample_leaf represents the minimum number of samples in the leaf nodes of the decision trees. Usual values to min_sample_leaf are 1, 3, 5, 10, 25, ..

Doing so has the following benefits:

* There are less decisions to make in leaf node: should generalize better
* The predictions are made by averaging more samples in leaf level: reducing volatility of decisions

```
reset_rf_samples()
```

```
for i in [1,3,5]:
  m = RandomForestRegressor(n_estimators=40, n_jobs=-1, min_samples_leaf=i)
  m.fit(X_train,y_train)
  print (i)
  print_score(m)
```

**max_feature**

We can also reduce the amount of correlation between the estimators by choosing a random sample of features along with the datapoints at each split. We can do this by specifying max_features. 

It usually takes the values 0.5, sqrt, log2, None(default)

```
m = RandomForestRegressor(n_estimators=40, n_jobs=-1, min_samples_leaf=3, max_features=0.5, oob_score=True)
%time m.fit(X_train,y_train)
print_score(m)
```

One simple way is to use the standard deviation of those predictions alongside the mean. What this gives us is relative confidence of predictions so that we can be more cautious in dealing with rows which have much variance.

####One Hot Encoding

One Hot Encoding is one of the techniques where we apply **Data Augmentation** on **Categorical** variables. Here we split a categorical feature based on its **cardinality** into an equal number of columns with a binary or boolean value to represent the categorical value of that data-point.

We do this by specifying max_n_cat in proc_df(). This splits every categorical column whose cardinality is lesser than or equal max_n_cat

Whether this improves our model or not is for us to find:

### Removing Redundant Features

One of the important pieces of Feature Engineering is to remove redundant features i.e features representing similar meanings or behavior.

We are going to be doing this by first finding features which have a similar nature. We do this by means of a **dendogram**. Dendogram is a kind of **Heirarchical Clustering**. It basically involves finding the closest pair of objects and replacing them with their average. We reduce the objects pairwise and can plot it.

```
from scipy.cluster import hierarchy as hc
```

We are going to be using something called **Spearman\`s R Correlation Coefficient**. It is a ranked correlation. Unlike R2, we use this ranked correlation coefficient to find the correlation between features or variables instead of predictions. The reason we choose Spearman\`t R in particular is because Random Forests never assumer linearity between the variables instead they base their decisions on the ordering.

```
corr = np.round(scipy.stats.spearmanr(df_keep).correlation, 4)
corr_condensed = hc.distance.squareform(1-corr)
z = hc.linkage(corr_condensed, method='average')
fig = plt.figure(figsize=(16,10))
dendrogram = hc.dendrogram(z, labels=df_keep.columns, orientation='left', leaf_font_size=16)
plt.show()
```

We can see the features that are closest on the leaf level on the dendogram represent features which have the most similar nature in measuring the same thing.

Once we infer from the dendogram, we can conclude the following features have similar nature:

*   saleYear & saleElapsed
*   Hydrailics_Flow & Grouser_Tracks and Coupler_Sytem
*   ProductGroup & ProductGroupDesc
*   fiBaseModel & fiModelDesc

We shall now test how removing each of these features affects the learning rate of our model.
We shall be using OOB Score as our metric since we need to genralize how removing a feature affects our model.



```
def get_oob(df):
    m = RandomForestRegressor(n_estimators=30, min_samples_leaf=5, max_features=0.6, n_jobs=-1, oob_score=True)
    x, _ = split_vals(df, n_trn)
    m.fit(x, y_train)
    return m.oob_score_
```

Our baseline OOB Score is:

```
get_oob(df_keep)
```

Now we remove each of the columns

```
for c in ('saleYear', 'saleElapsed', 'fiModelDesc', 'fiBaseModel', 'Grouser_Tracks', 'Coupler_System','ProductGroup','ProductGroupDesc'):
    print(c, get_oob(df_keep.drop(c, axis=1)))
```

We can see that the model less worse when we tried to tried to remove the following features compared to others:
* saleYear
* fiBaseModel
* Grouser_Tracks
* ProductGroup

Now we remove all of these redundant features from the dataset.
